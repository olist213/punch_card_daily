## API的结构

API端点是一个URL，可以帮助检索、创建、更新或删除资源。

API 使用以下 HTTP 方法进行对象操作：

```
GET   用于对象检索
POST  用于对象创建和对象操作
PUT  用于对象更新
DELETE  用于删除对象
```
