# api(Application Programming Interface)

API代表应用程序编程接口，我们将在本节学习web API，web API是定义的接口，通过这些接口，企业和使用其资产的应用程序之间发生互动，这也是一个服务级别协议（SLA），以指定服务提供者，并为其API用户公开服务路径或URL。

在网络开发的背景下，API被定义为一组规范，如http请求信息、响应信息，通常是xml或者是json格式。

网络API已经从基于简单对象访问协议（SOAP）的网络服务和面向服务的架构（SOA）转向更直接的表示状态传输（REST）风格的网络资源。

社会媒体服务、网络API已经允许网络社区在社区和不同平台之间分享内容和数据。

使用API，在一个地方动态创建的内容可以被发布和更新到网络上的多个位置。

例如，Twitter的REST API允许开发者访问Twitter的核心数据，搜索API为开发者提供了与Twitter搜索和趋势数据互动的方法。

许多应用程序提供API终端。
一些API的例子，如：
国家API（https://restcountries.eu/rest/v2/all）。
猫的品种API（https://api.thecatapi.com/v1/breeds）。

在本节中，我们将介绍一个RESTful API，它使用HTTP请求方法来获取、PUT、POST和DELETE数据。

# 建立API

RESTful API是一个应用程序接口（API），使用HTTP请求来GET、PUT、POST和DELETE数据。在前面的章节中，我们已经了解了python、flask和mongoDB。我们将利用我们获得的知识，使用Python flask和mongoDB数据库开发一个RESTful API。每个有CRUD（创建、读取、更新、删除）操作的应用程序都有一个API来创建数据、获取数据、更新数据或从数据库中删除数据。

要建立一个API，需要了解HTTP协议和HTTP请求和响应周期。

# HTTP（超文本传输协议）

HTTP是客户端和服务器之间的一个既定的通信协议。这里的客户端是指浏览器，服务器是指访问数据的地方。HTTP是一个网络协议，用于传递资源，这些资源可能是万维网上的文件，无论是HTML文件、图像文件、查询结果、脚本或其他文件类型。

浏览器是一个HTTP客户端，因为它向HTTP服务器（Web服务器）发送请求，然后由服务器向客户端发送响应。

# HTTP的结构

HTTP使用客户-服务器模式。一个HTTP客户端打开一个连接，并向HTTP服务器发送一个请求信息，HTTP服务器返回响应信息，这是请求的资源。当请求响应周期完成后，服务器关闭连接。



请求和响应信息的格式是相似的。

这两种消息都有：

- 初始行
- 零个或多个标题行。
- 一个空行（即一个CRLF本身
- 可选的body体

让我们通过浏览这个网站来了解请求和响应信息的例子：https://thirtydaysofpython-v1-final.herokuapp.com/。



# 初始请求行（状态行）

初始请求行与响应不同，请求行由三部分组成，以空格分隔：

- 方法名（GET、POST、HEAD）
- 请求资源的路径
- HTTP协议版本

GET是最常见的HTTP方法，有助于获取或读取资源，POST是创建资源的常见请求方法。

## 初始响应行（状态行）

初始化响应行，也由三部分组成。

- HTTP协议版本
- 响应状态码，以及迷哦啊叔状态代码的原因，http/1.0 200 ok或者http/1.0 404 Not Found

最常见的状态代码是。200 OK。请求成功，结果资源（如文件或脚本输出）将在消息正文中返回。
